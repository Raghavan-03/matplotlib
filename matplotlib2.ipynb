```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,':b>')
```

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='blue',marker='>',linestyle=":")
```

```python
plt.plot(x,y,color='black',marker='>',linestyle=':',markersize=10,alpha=0.5)
```

```python
plt.plot(x,y,'-k*',alpha=0.6)
```

```python
plt.plot(x,y,'r*',alpha=1) # alpha can be specified on a scale 0 to 1
```
